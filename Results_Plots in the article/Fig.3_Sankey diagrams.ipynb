```python
# Load and read the Excel data
file_path = "../Supporting Information 2_PlasticFADE.xlsx"  # CHECK: confirm file path
sheet_name = "Sankey"
df = pd.read_excel(file_path, sheet_name=sheet_name, usecols='K', skiprows=1)
values = df.iloc[:, 0].tolist()  # Convert the column data to a list of values
pd.set_option('display.float_format', '{:.3f}'.format)  # Display all floats in 3 decimal digits

df_PS_Beach = df.iloc[0:9]
df_PS_WSur = df.iloc[9:18]
print(df_PS_Beach, "\n", df_PS_WSur)
```

```python
# Sankey diagram for PS, Beach
# Manually add text to be shown on the plot
node_labels = ["Food containers: 100 kg", "Macroplastics: 2.5 kg", "Microplastics: 62.7 kg", "CO<sub>2</sub>+Organics: 34.8 kg", 
               "Macroplastics: 0.1 kg", "Microplastics: 42.5 kg", "CO<sub>2</sub>+Organics: 57.4 kg"]
node_colors = ['#00A1C0', '#0BAC43', '#FA961E', '#B91E32', '#0BAC43', '#FA961E', '#B91E32']

# Define sources, targets, and values for the Sankey diagram
source_nodes = [0, 0, 0, 1, 1, 1, 2, 2, 3]
target_nodes = [1, 2, 3, 4, 5, 6, 5, 6, 6]
link_values = df_PS_Beach.iloc[:, 0].values
np.set_printoptions(precision=3, suppress=True)
print(link_values)

# Calculate node values (sum of links for each node)
node_values = [0] * len(node_labels)
for i in range(len(source_nodes)):
    node_values[source_nodes[i]] += link_values[i]
    node_values[target_nodes[i]] += link_values[i]

# Format node labels with values (rounded to one decimal place and suffixed with 'kg')
formatted_labels = [f"{label}: {value:.1f} kg" for label, value in zip(node_labels, node_values)]

# Custom node positions: set 'x' and 'y' positions (adjust y for a better-looking plot)
x_positions = [0.06, 0.5, 0.5, 0.5, 0.94, 0.94, 0.94]  # x positions: left, middle, right
y_positions = [0.5, 0.04, 0.358, 0.824, 0.03, 0.25, 0.73]  # y positions: control vertical alignment (change manually)

# Create the Sankey diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        # pad=20,  # Space between nodes
        thickness=25,  # Node thickness
        label=node_labels,
        x=x_positions,
        y=y_positions,
        color=node_colors  # Node color
    ),
    link=dict(
        source=source_nodes,  # Source nodes
        target=target_nodes,  # Target nodes
        value=df_PS_Beach     # Link values
    )
)])

# Add annotations at positions 0.05, 0.5, 0.95
fig.add_annotation(
    x=0, y=-0.08,
    text="Plastic Emission",
    showarrow=False,
    font=dict(size=16),
    xref="paper", yref="paper"
)
fig.add_annotation(
    x=0.5, y=-0.08,
    text="After 5 years",
    showarrow=False,
    font=dict(size=16),
    xref="paper", yref="paper"
)
fig.add_annotation(
    x=1, y=-0.08,
    text="After 10 years",
    showarrow=False,
    font=dict(size=16),
    xref="paper", yref="paper"
)

# Update layout
fig.update_layout(
    title=dict(
        text="Mass transformation (PS, Beach, k<sub>frag</sub> > k<sub>degr</sub>)",
        x=0.5,
        xanchor='center',
        font=dict(size=18)
    ),
    font=dict(size=15, color="black"),
    width=1000,  # Set width in pixels
    height=500,  # Set height in pixels
    margin=dict(l=15, r=15, t=60, b=60), # Adjust bottom margin for annotations
)

fig.add_annotation(x=0, y=1.13, text="a)", showarrow=False, font=dict(size=20, weight="normal"), xref="paper", yref="paper")
fig.write_image("../Figures/Figure_3a.png", width=1000, height=500, scale=6)  # CHECK: confirm name and desired place for the plot
fig.show()
```

```python
# Load and read the Excel data
import pandas as pd
import plotly.graph_objects as go
from PIL import Image
import os

file_path = "../Supporting Information 2_PlasticFADE.xlsx"
sheet_name = "Sankey"
df = pd.read_excel(file_path, sheet_name=sheet_name, usecols='K', skiprows=1)
pd.set_option('display.float_format', '{:.3f}'.format)

df_PS_Beach = df.iloc[0:9]
df_PS_WSur = df.iloc[9:18]

# =============================================
# FIRST SANKEY: PS Beach (exactly as original)
# =============================================
node_labels_1 = ["Food containers: 100 kg", "Macroplastics: 2.5 kg", "Microplastics: 62.7 kg", "CO<sub>2</sub>+Organics: 34.8 kg", 
                 "Macroplastics: 0.1 kg", "Microplastics: 42.5 kg", "CO<sub>2</sub>+Organics: 57.4 kg"]
node_colors_1 = ['#00A1C0', '#0BAC43', '#FA961E', '#B91E32', '#0BAC43', '#FA961E', '#B91E32']

source_nodes_1 = [0, 0, 0, 1, 1, 1, 2, 2, 3]
target_nodes_1 = [1, 2, 3, 4, 5, 6, 5, 6, 6]
link_values_1 = df_PS_Beach.iloc[:, 0].values

x_positions_1 = [0.06, 0.5, 0.5, 0.5, 0.94, 0.94, 0.94]
y_positions_1 = [0.5, 0.04, 0.358, 0.824, 0.03, 0.25, 0.73]

fig1 = go.Figure(data=[go.Sankey(
    node=dict(
        thickness=25,
        label=node_labels_1,
        x=x_positions_1,
        y=y_positions_1,
        color=node_colors_1
    ),
    link=dict(
        source=source_nodes_1,
        target=target_nodes_1,
        value=link_values_1
    )
)])

fig1.update_layout(
    title=dict(
        text="Mass transformation (PS, Beach, k<sub>frag</sub> > k<sub>degr</sub>)",
        x=0.5,
        xanchor='center',
        font=dict(size=18)
    ),
    font=dict(size=15, color="black"),
    width=1000,
    height=500,
    margin=dict(l=15, r=15, t=60, b=60),
)

fig1.add_annotation(x=0, y=-0.08, text="Plastic Emission", showarrow=False, font=dict(size=16), xref="paper", yref="paper")
fig1.add_annotation(x=0.5, y=-0.08, text="After 5 years", showarrow=False, font=dict(size=16), xref="paper", yref="paper")
fig1.add_annotation(x=1, y=-0.08, text="After 10 years", showarrow=False, font=dict(size=16), xref="paper", yref="paper")
fig1.add_annotation(x=0, y=1.13, text="a)", showarrow=False, font=dict(size=20, weight="normal"), xref="paper", yref="paper")

# =============================================
# SECOND SANKEY: PS Water surface (exactly as original)
# =============================================
node_labels_2 = ["Food containers: 100 kg", "Macroplastics: 82.0 kg", "Microplastics: 0.4 kg", "CO<sub>2</sub>+Organics: 17.6 kg", 
                 "Macroplastics: 67.3 kg", "Microplastics: 0.6 kg", "CO<sub>2</sub>+Organics: 32.1 kg"]
node_colors_2 = ['#00A1C0', '#0BAC43', '#FA961E', '#B91E32', '#0BAC43', '#FA961E', '#B91E32']

source_nodes_2 = [0, 0, 0, 1, 1, 1, 2, 2, 3]
target_nodes_2 = [1, 2, 3, 4, 5, 6, 5, 6, 6]
link_values_2 = df_PS_WSur.iloc[:, 0].values

x_positions_2 = [0.06, 0.5, 0.5, 0.5, 0.94, 0.94, 0.94]
y_positions_2 = [0.5, 0.4, 0.794, 0.896, 0.335, 0.67, 0.845]

fig2 = go.Figure(data=[go.Sankey(
    node=dict(
        thickness=25,
        label=node_labels_2,
        x=x_positions_2,
        y=y_positions_2,
        color=node_colors_2
    ),
    link=dict(
        source=source_nodes_2,
        target=target_nodes_2,
        value=link_values_2
    )
)])

fig2.update_layout(
    title=dict(
        text="Mass transformation (PS, Water surface, k<sub>frag</sub> << k<sub>degr</sub>)",
        x=0.5,
        xanchor='center',
        font=dict(size=18)
    ),
    font=dict(size=15, color="black"),
    width=1000,
    height=500,
    margin=dict(l=15, r=15, t=60, b=60),
)

fig2.add_annotation(x=0, y=-0.08, text="Plastic Emission", showarrow=False, font=dict(size=16), xref="paper", yref="paper")
fig2.add_annotation(x=0.5, y=-0.08, text="After 5 years", showarrow=False, font=dict(size=16), xref="paper", yref="paper")
fig2.add_annotation(x=1, y=-0.08, text="After 10 years", showarrow=False, font=dict(size=16), xref="paper", yref="paper")
fig2.add_annotation(x=0, y=1.13, text="b)", showarrow=False, font=dict(size=20, weight="normal"), xref="paper", yref="paper")

# Save both figures
fig1.write_image("temp_fig1.png", width=1000, height=500, scale=2)
fig2.write_image("temp_fig2.png", width=1000, height=500, scale=2)

# Combine images vertically
img1 = Image.open("temp_fig1.png")
img2 = Image.open("temp_fig2.png")

combined_img = Image.new('RGB', (img1.width, img1.height + img2.height))
combined_img.paste(img1, (0, 0))
combined_img.paste(img2, (0, img1.height))

combined_img.save("../Figures/Figure_3_combined.png", dpi=(600, 600))
combined_img.show()

# Clean up
os.remove("temp_fig1.png")
os.remove("temp_fig2.png")
```
