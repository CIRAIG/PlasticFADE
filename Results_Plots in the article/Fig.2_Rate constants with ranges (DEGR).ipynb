```python
import pandas as pd
# Load data from the Excel sheet
file_path = "../Supporting Information 2_PlasticFADE.xlsx"  # CHECK: confirm file path
sheet_name = "Rate constants"

# Read and display the data
data_D = pd.read_excel(file_path, sheet_name=sheet_name, usecols='F:J', skiprows=2)
data_D_1 = data_D.iloc[0:21, :]
pd.set_option('display.float_format', '{:.2e}'.format)
print(data_D_1)
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator
import numpy as np

# Set up the figure and axis
plt.figure(figsize=(14, 5))

data_D_1 = data_D_1.copy()
x = range(len(data_D_1))
data_D_1['x_labels'] = data_D_1['Polymer.1'] + '\n' + data_D_1['Compartment.1']

y = data_D_1['kdegr [d-1]']
yerr_lower = y - data_D_1['kdegr_LL[d-1]']
yerr_upper = data_D_1['kdegr_UL[d-1]'] - y
yerr = [yerr_lower, yerr_upper]  # Asymmetric error bars

# Specify indices for each color-coding category (0-based)
idx_low = [2, 5, 6, 9, 12, 13]
idx_moderate = [8]
idx_high = [i for i in range(len(data_D_1)) if i not in idx_low + idx_moderate]

# Plot main results with asymmetric error bars
def plot_points(indices, color, label):
    first = True
    for i in indices:
        yi  = y.iloc[i]
        yerr_i = [[yerr_lower.iloc[i]], [yerr_upper.iloc[i]]]
        plt.errorbar(
            [i], [yi], yerr=yerr_i,
            fmt='o',
            color=color,  # marker color
            ecolor=color,  # error bar color (use 'black' here if you prefer black whiskers)
            markersize=5,
            elinewidth=1,
            capsize=4,
            label=label if first else None
        )
        first = False

plot_points(idx_high, '#0BAC43', 'Rate constants (high level of confidence)')
plot_points(idx_moderate, '#FA961E', 'Rate constants (moderate level of confidence)')
plot_points(idx_low, '#B91E32', 'Rate constants (low level of confidence)')

# Indices (0-based) to ALWAYS show the bottom "0" arrow for
force_zero_idx = [2, 5, 6, 9, 12, 13]

def annotate_bottom_zero(i, color='#8c8c8c'):
    y_zero_marker = min_display_y * 1
    y_zero_label  = y_zero_marker * 2
    plt.annotate(
        '(0)',
        xy=(i, y_zero_marker),
        xytext=(i, y_zero_label),
        textcoords='data',
        ha='center',
        va='bottom',
        fontsize=13,
        color=color,
        arrowprops=dict(arrowstyle='-|>', color=color, lw=1),
        clip_on=False,
    )

# Show the outliers (outside the plot boundary)
y_max = 1e-1
label_y = y_max * 10**0.25  # X% above the plot top
min_display_y = 1e-9

for i in range(len(data_D_1)):
    upper = data_D_1['kdegr_UL[d-1]'].iloc[i]
    if upper > y_max:
        label = f"{upper:.1e}".replace("+", "")

        plt.plot(
            [i, i],              # x-coordinates
            [y_max, label_y],    # from plot top to label
            color='grey',
            linewidth=1,
            linestyle=':',
            clip_on=False  
        )       

    # Draw text label
        plt.text(
            i, label_y,
            label,
            ha='center',
            va='bottom',
            # fontsize=9,
            color='grey'
        )

# Annotate 0 and arrows if i is in the forced list
for i in range(len(data_D_1)):
    if i in force_zero_idx:
        annotate_bottom_zero(i)

# Configure x-axis
plt.xticks(x, data_D_1['x_labels'])
plt.yscale('log')
plt.xlabel('Polymer & Compartment', labelpad=10, fontsize=12)
plt.ylabel('Degradation rate constant [d$^{-1}$]', labelpad=5, fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=11)
plt.legend(loc='upper right', handletextpad=0.2, fontsize=11)
plt.xlim(-0.5, len(data_D_1) - 0.5)  # Pull the first and last data points closer to the edge of the plot
plt.ylim(1e-9, 1e-1)  # Set the y-axis range (change manually)
plt.gca().yaxis.set_major_locator(FixedLocator([10**exp for exp in range(-9, 0, 1)]))  # Manually set the tick positions at every 1e1
plt.grid(True, which="major", ls="--", lw=0.5)
plt.text(-0.07, 1.0, "a)", transform=plt.gca().transAxes, fontsize=15, fontweight='normal')
plt.savefig("../Figures/Figure_2a.png", dpi=600, bbox_inches='tight')  # CHECK: confirm name and desired place for the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator
import numpy as np

# Create figure with two subplots stacked vertically
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

# =============================================
# TOP SUBPLOT (a)
# =============================================
data_D_1 = data_D_1.copy()
x1 = range(len(data_D_1))
data_D_1['x_labels'] = data_D_1['Polymer.1'] + '\n' + data_D_1['Compartment.1']

y1 = data_D_1['kdegr [d-1]']
yerr_lower1 = y1 - data_D_1['kdegr_LL[d-1]']
yerr_upper1 = data_D_1['kdegr_UL[d-1]'] - y1

# Specify indices for each color-coding category
idx_low1 = [2, 5, 6, 9, 12, 13]
idx_moderate1 = [8]
idx_high1 = [i for i in range(len(data_D_1)) if i not in idx_low1 + idx_moderate1]

# Plot points
def plot_points(ax, indices, color, label, y, yerr_lower, yerr_upper, first_label=True):
    first = True
    for i in indices:
        yi = y.iloc[i]
        yerr_i = [[yerr_lower.iloc[i]], [yerr_upper.iloc[i]]]
        ax.errorbar([i], [yi], yerr=yerr_i, fmt='o', color=color, ecolor=color,
                   markersize=5, elinewidth=1, capsize=4,
                   label=label if (first and first_label) else None)
        first = False

plot_points(ax1, idx_high1, '#0BAC43', 'Rate constants (high level of confidence)', y1, yerr_lower1, yerr_upper1)
plot_points(ax1, idx_moderate1, '#FA961E', 'Rate constants (moderate level of confidence)', y1, yerr_lower1, yerr_upper1)
plot_points(ax1, idx_low1, '#B91E32', 'Rate constants (low level of confidence)', y1, yerr_lower1, yerr_upper1)

# Outliers and zero annotations
y_max = 1e-1
label_y = y_max * 10**0.25
min_display_y = 1e-9
force_zero_idx1 = [2, 5, 6, 9, 12, 13]

def annotate_bottom_zero(ax, i, color='#8c8c8c'):
    y_zero_marker = min_display_y * 1
    y_zero_label = y_zero_marker * 2
    ax.annotate('(0)', xy=(i, y_zero_marker), xytext=(i, y_zero_label),
               textcoords='data', ha='center', va='bottom', fontsize=13,
               color=color, arrowprops=dict(arrowstyle='-|>', color=color, lw=1), clip_on=False)

for i in range(len(data_D_1)):
    upper = data_D_1['kdegr_UL[d-1]'].iloc[i]
    if upper > y_max:
        label = f"{upper:.1e}".replace("+", "")
        ax1.plot([i, i], [y_max, label_y], color='grey', linewidth=1, linestyle=':', clip_on=False)
        ax1.text(i, label_y, label, ha='center', va='bottom', color='grey')
    if i in force_zero_idx1:
        annotate_bottom_zero(ax1, i)

# Formatting
ax1.set_xticks(x1)
ax1.set_xticklabels(data_D_1['x_labels'])
ax1.set_yscale('log')
ax1.set_ylabel('Degradation rate constant [d$^{-1}$]', labelpad=5, fontsize=12)
ax1.tick_params(axis='both', which='major', labelsize=11)
ax1.set_xlim(-0.5, len(data_D_1) - 0.5)
ax1.set_ylim(1e-9, 1e-1)
ax1.yaxis.set_major_locator(FixedLocator([10**exp for exp in range(-9, 0, 1)]))
ax1.grid(True, which="major", ls="--", lw=0.5)
ax1.text(-0.07, 1.0, "a)", transform=ax1.transAxes, fontsize=15, fontweight='normal')

# =============================================
# BOTTOM SUBPLOT (b)
# =============================================
data_D_2 = data_D_2.copy()
x2 = range(len(data_D_2))
data_D_2['x_labels'] = data_D_2['Polymer.1'] + '\n' + data_D_2['Compartment.1']

y2 = data_D_2['kdegr [d-1]']
yerr_lower2 = y2 - data_D_2['kdegr_LL[d-1]']
yerr_upper2 = data_D_2['kdegr_UL[d-1]'] - y2

idx_high2 = [14, 15, 17, 18]
idx_low2 = [2, 5, 6, 9, 12, 13, 16, 19, 20]
idx_moderate2 = [i for i in range(len(data_D_2)) if i not in idx_high2 + idx_low2]

plot_points(ax2, idx_high2, '#0BAC43', 'Rate constants (high level of confidence)', y2, yerr_lower2, yerr_upper2, first_label=False)
plot_points(ax2, idx_moderate2, '#FA961E', 'Rate constants (moderate level of confidence)', y2, yerr_lower2, yerr_upper2, first_label=False)
plot_points(ax2, idx_low2, '#B91E32', 'Rate constants (low level of confidence)', y2, yerr_lower2, yerr_upper2, first_label=False)

force_zero_idx2 = [2, 5, 6, 9, 12, 13, 16, 19, 20]

for i in range(len(data_D_2)):
    upper = data_D_2['kdegr_UL[d-1]'].iloc[i]
    if upper > y_max:
        label = f"{upper:.1e}".replace("+", "")
        ax2.plot([i, i], [y_max, label_y], color='grey', linewidth=1, linestyle=':', clip_on=False)
        ax2.text(i, label_y, label, ha='center', va='bottom', color='grey')
    if i in force_zero_idx2:
        annotate_bottom_zero(ax2, i)

# Formatting
ax2.set_xticks(x2)
ax2.set_xticklabels(data_D_2['x_labels'])
ax2.set_yscale('log')
ax2.set_xlabel('Polymer & Compartment', labelpad=10, fontsize=12)
ax2.set_ylabel('Degradation rate constant [d$^{-1}$]', labelpad=5, fontsize=12)
ax2.tick_params(axis='both', which='major', labelsize=11)
ax2.set_xlim(-0.5, len(data_D_2) - 0.5)
ax2.set_ylim(1e-9, 1e-1)
ax2.yaxis.set_major_locator(FixedLocator([10**exp for exp in range(-9, 0, 1)]))
ax2.grid(True, which="major", ls="--", lw=0.5)
ax2.text(-0.07, 1.0, "b)", transform=ax2.transAxes, fontsize=15, fontweight='normal')

# =============================================
# Single legend (placed below both subplots)
# =============================================
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(loc='upper right', handletextpad=0.2, fontsize=11)
#fig.legend(handles, labels, loc='lower center', ncol=3, handletextpad=0.2, fontsize=11, bbox_to_anchor=(0.5, 0.02))

plt.tight_layout()
plt.subplots_adjust(bottom=0.1)  # Make space for legend
plt.savefig("../Figures/Figure_2_combined.png", dpi=600, bbox_inches='tight')
plt.show()
```
